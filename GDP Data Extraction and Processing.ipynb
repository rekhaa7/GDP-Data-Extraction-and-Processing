```python
import numpy as np
import pandas as pd
```

```python
URL="https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
```

```python
# Extracting tables frim weboage using Pandas

tables = pd.read_html(URL)
df = tables[3]
df
```

```python
# Replacing the column headers with column numbers

df.columns = range(df.shape[1])
df
```

```python
# Getting columns with index 0 to 2
df = df[[0,2]]

# Getting the rows with index 1 to 10, indicating the to 10 economies of the world
df = df.iloc[1:11,:]
df
```

```python
# Assign column names as "Country" and "GDP (Million GDP)"
df.columns = ['Country', 'GDP (Million USD)']
df
```

```python
# Changing the data type of the 'GDP (Million USD)' column to integer
df['GDP (Million USD)'] = df['GDP (Million USD)'].astype(int)

# Converting the GDP value in million USD to billion USD
df[['GDP (Million USD)']] = df[['GDP (Million USD)']]/1000

# Using numpy.round() method to round the value to 2 decimal places
df[['GDP (Million USD)']] = np.round(df[['GDP (Million USD)']], 2)

# Renaming the column header from 'GDP (Million USD)' to 'GDP (Billion USD)'
df.rename(columns = {'GDP (Million USD)' : 'GDP (Billion USD)'})
```

```python
# Loading the DataFrame to the CSV file named "Largest_economies.csv"
df.to_csv('./Largest_economies.csv')
```
